```python
import numpy as np
import matplotlib.pyplot as plt

from traffic_signs import load_data

import tensorflow as tf
from sklearn.model_selection import train_test_split
#from keras.models import Sequential
#from keras.models import Model
#from keras.layers import Dense
#from keras.layers import Conv2D
#from keras.layers import MaxPooling2D
#from keras.layers import Flatten
```

## Load data

```python
X, y = load_data(".")
```

```python
def one_hot(data, nb_classes):
    targets = np.array(data).reshape(-1)
    return np.eye(nb_classes)[targets]
```

```python
X = X.astype(np.float32) / 255

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

X_train = X_train[..., np.newaxis]
X_test = X_test[..., np.newaxis]
y_train = one_hot(y_train, 2)
y_test = one_hot(y_test, 2)

# 1) Load X and y.
X = np.loadtxt('X.csv', delimiter=',')
y = np.loadtxt('y.csv')      
print(X)
print(y)
```

```python
print(X_train[0].shape)
```

```python
model = tf.keras.models.Sequential()
```

```python
# model architecture
N = 32
w, h = 5, 5

# change input shape 3rd value 3 -> 1
model.add(tf.keras.layers.Conv2D(N, (w, h), input_shape=(64, 64, 1), activation = 'relu', padding = 'same'))
model.add(tf.keras.layers.MaxPooling2D(pool_size=(4, 4))) # pool size 2 -> 4
model.add(tf.keras.layers.Conv2D(N, (w, h), activation = 'relu', padding = 'same'))
model.add(tf.keras.layers.MaxPooling2D((4,4)))                  # add this
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(units = 100, activation = 'sigmoid'))   # add this
model.add(tf.keras.layers.Dense(units = 2, activation = 'sigmoid'))         
```

```python
model.summary()
```

```python
# compiling model
model.compile(loss='categorical_crossentropy', optimizer='sgd', metrics=['accuracy'])

# training model, validation_data = [X_test, y_test] = optimizer will report the test error every epoch
model.fit(X_train, y_train, batch_size=32, epochs=20, validation_data = [X_test, y_test] )
```
