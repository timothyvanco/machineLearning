Train classifiers for the GTSRB task

```python
import numpy as np
import matplotlib.pyplot as plt
from simplelbp import local_binary_pattern
from traffic_signs import load_data
from traffic_signs import extract_lbp_features
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
```

### Extract Data

```python
X, y = load_data(".")
F = extract_lbp_features(X)
```

### Test different Classifier
Evaluation of the performance with cross validation score

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
```

```python
knn = KNeighborsClassifier()
lda = LinearDiscriminantAnalysis()
svm = SVC()
logreg = LogisticRegression()
classifiers = [knn, lda, svm, logreg]
for c in classifiers:
    acc = cross_val_score(c,F, y)
    print(acc)
    
```
