```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
w = np.sqrt(0.3) * np.random.randn(100)
f = 0.015
n = np.linspace(0,100,100)

x = np.sin(2*math.pi*f*n)+w
plt.plot(x)


# x[n] = sin (2πfn) + w[n]
```

```python
scores = [] 
frequencies = []

for f in np.linspace(0, 0.5, 1000):
    
    # Create vector e. Assume data is in x.
    
    n = np.arange(100)
    z = -2*math.pi*1j*f*n         # <compute -2*pi*i*f*n. Imaginary unit is 1j> e = numpy.exp(z)
    e = np.exp(z)
    score = np.abs(np.dot(x, e))          # <compute abs of dot product of x and e> scores.append(score)
    scores.append(score)
    frequencies.append(f)
    

fHat = frequencies[np.argmax(scores)]
print(fHat)
```
